```python
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')
```

```python
# sentiment-analysis: 긍부정 판별

classifier=pipeline('sentiment-analysis')
classifier(["I've been waiting for a Hugging Face course my whole life.", "I hate this so much!"])
```

```python
# zero-shot-classification: 주어진 label과의 연관성

classifier=pipeline('zero-shot-classification')
classifier("This is a course about the Transformers library",
           candidate_labels=['education','politics','buisiness'])
```

```python
# text-generation: 이후 문장 예측

generator=pipeline("text-generation",model='gpt2')
generator("In this course, we will teach you how to",
          max_new_tokens=50
          )
```

```python
# fill-mask: 문장 속 빈 단어 예측

unmasker=pipeline('fill-mask')
unmasker('This course will teach you all about <mask> models.',top_k=2)
```

```python
# NER: 문장 속 단어 정보 예측

ner=pipeline("ner",aggregation_strategy='simple')
ner("My name is Slyvain and I work at Hugging Face in Brooklyn.")
```

```python
# question-answering: context를 통해 question에 해당하는 답 예측

question_answerer=pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn"
)
```

```python
# image-classification: 이미지 인식

image_classifier=pipeline(
    task='image-classification',model='google/vit-base-patch16-224'
)
result=image_classifier(
  "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/pipeline-cat-chonk.jpeg"
)
print(result)
```
